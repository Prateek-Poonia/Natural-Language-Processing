```python
from transformers import BartTokenizer, BartForConditionalGeneration
```

```python
model_name = 'facebook/bart-large-cnn'
tokenizer = BartTokenizer.from_pretrained(model_name)
model = BartForConditionalGeneration.from_pretrained(model_name)
#
```

```python
text = """
Artificial Intelligence and Machine Learning are transforming industries and the way businesses operate
 by automating processes, enhancing decision-making, and unlocking new insights from data.
  These technologies are not only optimizing operations but also enabling companies
   to offer more personalized and efficient services.
   In healthcare, AI and ML assist in early diagnosis and personalized treatment plans,
    improving patient outcomes. In finance, they power fraud detection, risk assessment,
     and algorithmic trading, making transactions faster and safer. Manufacturing industries
      are leveraging AI-driven automation for quality control and predictive maintenance,
       reducing downtime and increasing productivity. From retail to agriculture,
       these technologies are revolutionizing traditional practices, creating smarter systems,
        and providing competitive advantages that were once unimaginable.
        As a result, businesses that adopt AI and ML are better positioned to innovate and adapt in a rapidly evolving digital landscape
 """
```

```python
inputs = tokenizer(text, return_tensors='pt', max_length=1024, truncation=True)
```

```python
summary_ids = model.generate(inputs['input_ids'], num_beams=4, max_length=100,length_penalty=2.0,min_length=30, early_stopping=True)
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
```

```python
print("Original Text:")
print(text)
print("\n")
print("Summary:")
print(summary)
```
